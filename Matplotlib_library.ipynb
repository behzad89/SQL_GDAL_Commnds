```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import statsmodels.api as sm

df = sm.datasets.get_rdataset('GaltonFamilies', package='HistData').data
df.head(3)
```

```python
# To plot the columns of the df:

# 1. the columns' name should be exported as list
#clm_names = list(df)
clm_names = ['father','mother','childHeight','midparentHeight']

# 2. to plot the plot in the same ax:
fig,ax = plt.subplots()

for i in range(0,2): # end:The number of the columns
    ax.plot(df['family'], df[clm_names[i]], linestyle=':')

plt.show()


# to plot the two axes:

fig,axes = plt.subplots(2,1, sharey=True)

for i, ax in enumerate(axes.flat):
    ax.plot(df['family'], df[clm_names[i]])
    
plt.show()

# to plot each column in the different axes:
fig, axes = plt.subplots(2,2)
for i, ax in enumerate(axes.flat):
    ax.plot(df['family'], df[clm_names[i]])
    ax.set_title(clm_names[i])
    ax.set_ylabel('Height')
plt.show()
```
