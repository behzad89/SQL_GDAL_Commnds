```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

import os
import sys
import warnings
```

```python
plt.style.use("bmh")
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.species.unique()
```

# Univariate Analysis

```python
sns.scatterplot(x=data.index, y=data.sepal_length,hue=data.species)
plt.show()
```

```python
feat_col = data.columns[:-1]
feat_col
```

```python
fig, axes = plt.subplots(1,4,figsize=(20,6),sharey=True)
for ax, feat in zip(axes.flatten(),feat_col):
    sns.histplot(data, x=feat,bins=18,ax=ax,kde=True)
plt.show()
```

# BIvariate Analysis

```python
sns.FacetGrid(data,hue='species',height=5).map(sns.scatterplot,"sepal_length",'sepal_width').add_legend()
plt.show()
```

# Multivariate Analysis

```python
g = sns.PairGrid(data,hue="species")
g.map_diag(sns.histplot,hue=None,kde=True)
g.map_offdiag(sns.scatterplot)
plt.show()
```

```python
sns.pairplot(data,hue='species',height=3)
plt.show()
```
