# 1st Graph
* One the easiest way to plot is the `plot` function. It can have various variables to make the graph looks nice. In the following page you can find more info. regarding the variables:\
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
* `figsize` is based on the inch

```python
fig = plt.figure(1,figsize=(5,5))
plt.plot(x_data,y_data,
            marker='o',
            markersize=5,
            mfc='white',
            linestyle='--',
            alpha=0.8,
            color='m',
            label='data set 1')

plt.legend(loc='upper right')
plt.show()
```

# 2nd Graph
* To add two graphs in the same plot, it is regured to define the figure first, then it is possible to add them in the same figure.
* The difference between `semilogy/semilogx` and `plot` is that the 1st one applies `log` to the data
* Using `$` sign makes it possible to add mathematical formula as label

```python
fig = plt.figure(1,figsize=(5,5))
plt.semilogy(x_data,y_data,
            marker='o',
            markersize=8,
            mfc='white',
            linestyle='--',
            alpha=0.8,
            label='$y=x^2$')

plt.semilogy(x_data,y_data+500,
            marker='s',
            markersize=8,
            mfc='white',
            linestyle='--',
            alpha=0.8,
            label='$y=x^2+500$')

plt.xlabel('x variable')
plt.ylabel('y variable')
plt.legend(loc='lower right')
plt.show()
```
